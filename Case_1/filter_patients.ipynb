## Tokenize the Symptom into lists

```python
data_patient['Symptom_List'] = data_patient['symptoms'].str.split(',').apply(lambda toks: [t.strip() for t in toks])
```

## Filters patients older than 40 with more than 2 symptoms

```python
data_patient[(data_patient['age'] > 40) & (data_patient['Symptom_List'].apply(len) > 2)]
```
